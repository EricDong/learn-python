```python
from operator import itemgetter
    

# 读取文件,生成学生成绩表,一维列表
with open('report.txt') as f:
    scores = f.readlines()


#学生成绩的二维数组
stu_score = []
for score in scores:
    stu_score.append(score.split())
# print(stu_score)
# print(type(score))

# 取出科目title
subject = stu_score[0][1:]
# print(subject)

# 科目数量,学生数量
sub_count = len(subject)
stu_count = len(stu_score) -1

# print(stu_count)
# print(sub_count)

print('共有%d学生,每位学生有%d门课程'%(stu_count,sub_count))




# 计算每位学生的总成绩和平均成绩
for i in range(1,len(stu_score)):
    sum = 0
    for j in range(1,len(stu_score[i])):
        sum += int(stu_score[i][j])
    stu_score[i].append(sum)
    stu_score[i].append(round(sum/sub_count,1))


stu_score[0].append('总分')
stu_score[0].append('平均分')
# print(stu_score)
# print(len(stu_score[0]))  
    


# 计算每个科目和总分的平均成绩
total = []
total.append('平均')
# print(total)

for j in range(1,len(stu_score[0])):
    sub_sum = 0
    for i in range(1,len(stu_score)):
        sub_sum += int(stu_score[i][j])
    total.append(round(sub_sum/stu_count,1))
# print(total)
    
stu_score.insert(1,total)
# print(stu_score)

# print(type(stu_score[1][10]))
# print(stu_score[1][10])


# 根据总成绩从高到低重新排序,并添加名次
grade = stu_score[2:]
# print(grade)

grade = sorted(grade, key = itemgetter(10), reverse = True)
# print(grade)

for i in range(len(grade)):
    grade[i].insert(0,i+1)
# print(grade)

total.insert(0,0)
stu_score[0].insert(0,'名次')
grade.insert(0,total)
grade.insert(0,stu_score[0])
# print(grade)





# 替换60分以下的成绩为不及格
for m in range(2,len(grade)):
    for n in range(2,len(grade[1])):
        if int(grade[m][n]) < 60:
            grade[m][n] = '不及格'

print(grade)


# 处理后的成绩另存为新文件(result.txt)


with open('result.txt','w',encoding = 'utf-8') as f:
    for line in grade:
        f.write(' '.join('%s' %id for id in line)+'\n')

    
    
```
